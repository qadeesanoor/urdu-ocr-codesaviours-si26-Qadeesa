## Part B — Step 5: Research Task

**What is OCR (Optical Character Recognition)?**

OCR is a technology that converts images of printed or handwritten text into machine-readable text. It works by detecting shapes in an image, matching them to known character patterns, and outputting the corresponding text characters. This is what lets a scanned document or a photo of a sign become searchable, editable text instead of just a picture.

**Why is Urdu OCR harder than English OCR?**

Urdu is written in the Nastaliq script, which is cursive and connects letters differently depending on their position in a word — unlike English's fixed, separated letterforms. Urdu is also written right-to-left, letters change shape based on what comes before and after them (contextual forms), and diacritical marks (zer, zabar, pesh) can be added or omitted without changing the base letter. There is also far less labeled training data available for Urdu compared to English, which makes model training harder.

**Two real-world situations where Urdu OCR would be useful:**

1. Digitizing old Urdu newspapers, government records, and books so they become searchable and preservable instead of degrading physical copies.
2. Automatically extracting text from Urdu signboards, ID documents, or handwritten forms — useful for apps that translate signs for travelers, or for automating data entry from Urdu paperwork.

```python
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display
import os

# Sample Urdu sentences to generate
urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے",
    "علم انسان کو روشنی دیتا ہے",
    "پانی زندگی کی بنیاد ہے",
    "وقت کی قدر کرنی چاہیے",
    "دوستی ایک انمول رشتہ ہے",
    "محنت میں عظمت ہے",
    "کتاب علم کا خزانہ ہے",
    "صحت ہزار نعمت ہے",
    "سچائی ہمیشہ کامیاب ہوتی ہے",
    "لاہور پاکستان کا ثقافتی مرکز ہے",
    "بارش کے بعد موسم سہانا ہو گیا",
    "استاد قوم کا معمار ہوتا ہے",
    "غریبوں کی مدد کرنی چاہیے",
    "ہر مشکل کے بعد آسانی ہے",
    "انسان اپنی محنت سے پہچانا جاتا ہے",
    "قومی زبان کی اہمیت سب کو سمجھنی چاہیے",
    "طلبہ کو وقت کی پابندی کرنی چاہیے",
    "درخت لگانا ماحول کے لیے ضروری ہے",
]

os.makedirs('data/raw/synthetic', exist_ok=True)

FONT_PATH = "NotoNastaliqUrdu-Regular.ttf"
FONT_SIZE = 32

try:
    font = ImageFont.truetype(FONT_PATH, FONT_SIZE)
except OSError:
    print(f"WARNING: font not found at {FONT_PATH}. Download a Nastaliq font before running this for real.")
    font = ImageFont.load_default()

for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)
    img = Image.new('RGB', (500, 90), color='white')
    draw = ImageDraw.Draw(img)
    draw.text((20, 25), bidi_text, fill='black', font=font)
    img.save(f'data/raw/synthetic/urdu_{i+1:03d}.png')
    print(f'Generated: urdu_{i+1:03d}.png -- {text}')

print('Done.')
```
